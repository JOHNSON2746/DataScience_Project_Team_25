```python
import pandas as pd
import numpy as np
import random
```

```python
# Total number of smaple
n_total = 100
```

```python
# proportion of male and female
p_male_ratio = 0.49
p_female_ratio = 0.51

# compute num of male and female 
n_male = int(n_total * p_male_ratio)
n_female = int(n_total * p_female_ratio)
```

```python
# Generate the Sex column
Sex = ['MALE'] * n_male + ['FEMALE'] * n_female
# shuffle the sex column 
np.random.shuffle(Sex)
```

```python
# Age group proportion based on 2021 dataset 

age_distribution = {
    "age_0_14": 18.3,
    "age_15_17": 3.42,
    "age_18_20": 2.28,
    "age_21_24": 6.2,
    "age_25_29": 7.0,
    "age_30_34": 7.3,
    "age_35_39": 7.0,
    "age_40_44": 6.5,
    "age_45_49": 6.2,
    "age_50_54": 6.3,
    "age_55_59": 6.1,
    "age_60_64": 5.9,
    "age_65_69": 5.1,
    "age_70_74": 4.6,
    "age_75_79": 3.2,
    "age_80_84": 2.2,
    "age_85_over": 2.1
}

for i in age_distribution.keys():
    age_distribution[i] = round(age_distribution[i] * 0.01, 4)

# Compute num of each age group
num_ageGroup_list = []

for i in age_distribution.keys():
    num_of_a_ageGroup = int(age_distribution[i] * n_total)
    num_ageGroup_list.append(num_of_a_ageGroup)

# Create a list for age label
age_labels = list(age_distribution.keys())

# Generate the Age column
Age = []
for label, count in zip(age_labels, num_ageGroup_list):
    Age.extend([label]*count)

# Check if the number of label value in column Age = n_total or not, if not, fill in the random number to reach the n_total
remaining = n_total - len(Age)
if remaining > 0:
    Age.extend(np.random.choice(age_labels, remaining))

# Shuffle the value in the Age column
np.random.shuffle(Age)
```

```python
syn_dataset  = pd.DataFrame({
    'Sex': Sex,
    'Age': Age
})
```

```python
syn_dataset["Median_income_per_year"] = syn_dataset.apply(
    lambda row: income_by_age_sex[row["Age"]][row["Sex"]],
    axis=1
)

# apply a random standard deviation to the income colum
pct_changes = [-0.15, -0.1, -0.08, -0.06, -0.04, -0.02, -0.015, -0.01, -0.005, 0, 0.005, 0.01, 0.015, 0.02, 0.04, 0.06, 0.08, 0.1, 0.15]
syn_dataset["Median_income_per_year"] = syn_dataset["Median_income_per_year"].apply(lambda x: x * (1 + np.random.choice(pct_changes)))
```

只用每个特征的分布和均值，不考虑联合分布。

初始化一批样本，只赋予收入数值，不赋特征标签。

对每个特征变量的每个类别，从样本池中选出一批（如800个），使得这批样本的平均收入尽量接近该类别的真实均值，并赋予该类别标签。

最终只保留那些被所有特征变量都赋值的样本（如2000个），作为生成数据。

```python
import pandas as pd
```

```python
# 输入源数据csv，输出该feature的收入分布
def compute_feature_income(df):
    group_col = 'feature'
    weight_col = 'number_of_jobs'
    income_col = 'median_income'
    prob = df.groupby(group_col)[weight_col].sum()
    prob = prob / prob.sum()
    def weighted_mean(x):
        return (x[weight_col] * x[income_col]).sum() / x[weight_col].sum()
    mean_income = df.groupby(group_col).apply(weighted_mean)
    stats = pd.DataFrame({
        'prob': prob,
        'mean_income': mean_income
    })
    return stats.dropna()
```

```python
csv_path = r"E:\git\DataScience_Project_Team_25\data\pre_transform\Industry_gender.csv"
df = pd.read_csv(csv_path)
compute_feature_income(df)
```

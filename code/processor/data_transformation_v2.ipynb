```python
# define a function to transofrm the median_income data from numeric to string(different bin of median income)
# <10k, 10K-20K, 20K-30k, 30k-50k, 50k-70k, 70k-90k, 90k-110k, 110k-130k, 130-150k, >150k
def median_income_transformation(datframe: pd.DataFrame, median_income_col: str):
    # define the bins and labels for median income
    bins = [0, 10000, 20000, 30000, 50000, 70000, 90000, 110000, 130000, 150000, float('inf')]
    labels = ['<10k', '10k-20k', '20k-30k', '30k-50k', '50k-70k', '70k-90k', '90k-110k', '110k-130k', '130-150k', '>150k']
    # apply the transformation to the median income column
    datframe[median_income_col] = pd.cut(datframe[median_income_col], bins=bins, labels=labels)
    return datframe


# define a function that can run the same above code for other similar csv files
def data_transformation(file_path: str, op_col1: str):
    # read the csv file
    df = read_csv(file_path, header=True)
    
    # change of string to uppercase in column "gender"
    df['gender'] = df['gender'].str.upper()

    # change the data type of the "median_income" column to numeric
    df['median_income'] = pd.to_numeric(df['median_income'], errors='coerce')

    # change values in column 'year' to string
    df['year'] = df['year'].astype(str)
    # apply the year transformation to the 'year' column
    df['year'] = df['year'].apply(year_transformation)
    # group by gender and the op_col1 (it could be 'Age)
    grouped_df = df.groupby(['gender', op_col1]).sum(numeric_only=True)
    total_jobs_by_sex = grouped_df.groupby('gender')['number_of_jobs'].sum()
    print(f'total_jobs_by_sex: {total_jobs_by_sex}.')
    # print the total number of 'PERSONS'
    total_persons = total_jobs_by_sex['PERSONS']
    print(f'Total number of persons: {total_persons}.')
    # inser a new column called 'PROPORTION' to the grouped dataframe
    grouped_df['PROPORTION'] = grouped_df.index.get_level_values('gender').map(total_jobs_by_sex)
    grouped_df['PROPORTION'] = grouped_df['number_of_jobs'] / total_persons  # grouped_df['PROPORTION']
    # filter out rows where 'gender' is 'PERSONS' or op_col1 is 'Industry' or 'Institutional sector'
    filter = (
        (grouped_df.index.get_level_values('gender') != 'PERSONS') &
        (grouped_df.index.get_level_values(op_col1) != 'Industry') &
        (grouped_df.index.get_level_values(op_col1) != 'Institutional sector') &
        (grouped_df.index.get_level_values(op_col1) != 'Type of legal organisation') & 
        (grouped_df.index.get_level_values(op_col1) != 'Employment size')
    )
    grouped_df = grouped_df[filter]

    # divided the median_income column by 5 
    if 'median_income' in grouped_df.columns:
        grouped_df['median_income'] = grouped_df['median_income'] / 5

    # transfrom the mdean income data from numeric to string
    if 'median_income' in grouped_df.columns:
        grouped_df = median_income_transformation(grouped_df, 'median_income')

    # show the edited dataframe as a dataframe format in a new line
    display(grouped_df)
    # print out the sum of the 'PROPORTION' column
    print(f'Sum of PROPORTION: {grouped_df["PROPORTION"].sum()}.')
    print('Data transformation completed successfully.')
    return grouped_df

# define a function for year transformation
# e.g.transform "2017-18" to "2017", "2018-2019 to "2018", "2019-20" to "2019...
def year_transformation(year_str):
    if '-' in year_str:
        year = year_str.split('-')[0]
    else:
        year = year_str
    return year

# read a csv file and set the first row as the header
# if header is False, the first row will be treated as data
def read_csv(file_path, header=True):
    if header:
        df = pd.read_csv(file_path)
    else:
        df = pd.read_csv(file_path, header=None)
    return df
```

# Process the transformed data from Table 17


```python
import pandas as pd

# define a function to convert the data column 'median_income' to numeric data column
# <10k, 10K-20K, 20K-30k, 30k-50k, 50k-70k, 70k-90k, 90k-110k, 110k-130k, 130-150k, >150k
def median_income_transformation(file_path, median_income_col):
    # Read the csv file
    df = pd.read_csv(file_path)
    # Define the bins and labels for median income
    bins = [0, 10000, 20000, 30000, 50000, 70000, 90000, 110000, 130000, 150000, float('inf')]
    labels = ['<10k', '10k-20k', '20k-30k', '30k-50k', '50k-70k', '70k-90k', '90k-110k', '110k-130k', '130-150k', '>150k']
    # Convert the median_income column to numeric, errors='coerce' will turn invalid parsing into NaN
    df[median_income_col] = pd.to_numeric(df[median_income_col], errors='coerce')
    # Apply the transformation to the median income column
    df[median_income_col] = pd.cut(df[median_income_col], bins=bins, labels=labels)
    # 合并同类项，基于 feature, gender, median_income 分组并汇总 number_of_jobs
    df = df.groupby(['feature', 'gender', median_income_col], as_index=False)['number_of_jobs'].sum()
    # compute the total number of jobs
    total_jobs_by_sex = df.groupby('gender')['number_of_jobs'].sum()
    print(f'total_jobs_by_sex: {total_jobs_by_sex}.')
    total_persons = total_jobs_by_sex['Persons']
    print(f'Total number of persons: {total_persons}.')

    # convert the values in number_of_jobs to numeric
    df['number_of_jobs'] = pd.to_numeric(df['number_of_jobs'], errors='coerce')

    # add a new column called "PROPORTION", which is the number_of_jobs divided by the sum of jobs based on combination of feature and gender
    df['PROPORTION'] = df['number_of_jobs'] / total_persons

    # remove the rows that gender is 'Persons'
    df = df[df['gender'] != 'Persons']
    return df
```
